```python
import pandas as pd
import numpy as np
```

## Reactions

```python
df = pd.read_csv('Reactions.csv')
df.head()
```

```python
df.shape
```

```python
# Removing unwanted column

df.drop(columns = ['Unnamed: 0','User ID'], inplace = True)
df.head()
```

```python
# Checking for null values

df.isna().sum()
```

```python
# removing null values

df = df.dropna()
df.head()
```

```python
df.shape
```

```python
Reaction = df
```

## Reaction Types

```python
df = pd.read_csv('ReactionTypes.csv')
df.head()
```

```python
df = df.drop(columns = 'Unnamed: 0')
df.head()
```

```python
df.shape
```

```python
ReactionTypes = df
```

## Content

```python
df = pd.read_csv('Content.csv')
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns = ['Unnamed: 0','User ID','URL'])
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
# Removing quotation marks in some values of category column

df['Category'] = df['Category'].str.strip(' " ')
```

```python
Content = df
Content.rename(columns = {'Type':'Content Type'}, inplace = True)
Content.head()
```

```python
Reaction.head()
```

```python
ReactionTypes.head()
```

```python
# Merging table Reaction and Content

Table = pd.merge(Reaction, Content, on = 'Content ID')
Table
```

```python
# Merging Final Table and Reaction Types

Table = pd.merge(Table, ReactionTypes, on = 'Type')
Table
```

```python
Table['Category'] = Table['Category'].str.capitalize()
Table.head()
```

```python
df = Table[['Category','Score']]
df
```

```python
Score = df.groupby('Category')['Score'].sum()
Score
```

```python
Score.sort_values(ascending = False)
```
